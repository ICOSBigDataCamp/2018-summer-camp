# 1. Word Count with Text
We start with a simple word count example where we count the word occurrences in a simple text file. Here, we use Lincoln's State of the Union address in 1861. Let's inspect the text file from command line.
## Word Count in Pure Python (without MapReduce)
This is how you would get the word counts in pure Python when the data are small enough to process from one machine.

````python
inpath="data/"
text_file="Lincoln_1861.txt" #Lincoln's State of the Union address in 1861

w_dict = {}
with open(inpath+text_file, 'r') as f1:
    for line in f1:
        for word in line.split():
            word = word.lower()
            if word not in w_dict:
                w_dict[word] = 1
            else:
                w_dict[word] += 1
````

````python
# Let's print the top 100 words in frequency of occurrence. Don't worry about the syntax for now.
s = [(k, w_dict[k]) for k in sorted(w_dict, key=w_dict.get, reverse=True)]
print(s[:100])
````

## Word Count with MapReduce
We import and use the `MRJob` class stored in the `mrjob` library.

In Python, a class provides a mechanism for bundling different functions together.
In our script, we define our own Python class that inherits the base `MRJob` class.
`MRJob` includes functions such as mapper(), reducer(), and run(), which are used to define and run your job.

````python
from mrjob.job import MRJob # We import the MRJob class which is the base class of the mrjob library
import json

class MRWordCounter(MRJob): # We define a class that "interits" the functions contained in the MRJob class.
    def mapper(self, key, line): # The first argument of a function inside a class is always "self". The mapper function takes a key and a value as inputs.
        for word in line.split():
            word = word.lower() # Force all characters to be lower case.
            
            ## The mapper outputs <key,value> pair. Here, word is the key, 1 is the value.
            yield word, 1  # yield is similar to "return", but can be called more than once. 

    def reducer(self, word, occurrences): # occurrences is an iterator that collects all the values of the same key (e.g. [1,1,1,1])
        yield word, sum(occurrences)

if __name__ == '__main__':  # If the Python interpreter executes this as an indepedent script rather than imported in another script, it will run this block.
    MRWordCounter.run() # Run the job that you implemented in MRWordCounter
````

We copy the above MRWordCounter class, save it as a python file (wordcount_mrjob.py), and run it from command line with an input text file (Lincoln_1861.txt) as below:
```
$ python wordcount_mrjob.py data/Lincoln_1861.txt
```

After executing both the mapper function and the reducer function, this command outputs on each line a unique word followed by its frequency of occurence as we instructed in our reducer function.

Sometimes it is useful for debugging purposes to see the output of just the mapper function, You can do this by adding `--mapper` to the above command:
```
$ python wordcount_mrjob.py data/Lincoln_1861.txt --mapper
```

To save the output to a text file, use the ">" sign followed by a file name:
```
$ python wordcount_mrjob.py data/Lincoln_1861.txt > Lincoln_1861_wordcount.txt
```

## Word Count in Pure Python (without MapReduce)
### This is how you would get the word counts in pure Python when the data are small enough to process from one machine.

````python
## Parse out just the tweet text from the data
texts = []
for tweet in all_tweets:
    texts.append( tweet['text'] )

print( "An example tweet text:" ) 
print(texts[0])
````

````python
## Create a Python dictionary where key is a word and value is its frequency
word_dict = {}
for one_text in texts: #iterate through each tweet
    words = one_text.split() # split the tweet into words (e.g words = ['Hello', 'world'])
    for w in words:
        w=w.lower() #force lowercase
        if w not in word_dict: #if the word is not in the dictionary yet
            word_dict[w] = 1 # put the word in the dictionary and set the value to 1
        else: # if the word is already in the dictionary
            word_dict[w] += 1 #increment the value by 1
````

### The Following code shows a MapReduce way of doing the Twitter word count.

````python
## Again, we import the MRJob class which is the base class of the mrjob library
import json
import mrjob
from mrjob.job import MRJob

class TwitterWordCount(MRJob):
    INPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #RawValueProtocol takes the entire line as the input value to the mapper. Assumes no key
    OUTPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #In the reducer, ignores the output key and only writes the value to file
    
    def mapper(self, _, line): # key is ignored in this example
        tweet = json.loads(line) # Convert raw string into Python object
        text = tweet['text'] # This is the tweet text (e.g. '@NeilECollins @nyc311 Thank you for reading, Mr. Collins.  Having a TLC driver license is a privilege.')
        
        ## Basic tokenization: Split the tweet text into words by whitespace
        words = text.split() #['@NeilECollins', '@nyc311', 'Thank', 'you', 'for', 'reading,', 'Mr.', 'Collins.', 'Having', 'a', 'TLC', 'driver', 'license', 'is', 'a', 'privilege.']
        for word in words:
            if word[0] != '@' and word[:4] != "http": # Filter out @mentions (e.g. @nyc311) and links (e.g. https://t.co....)
                '''
                Basic cleaning of each word: force all characters in a word to lower case. 
                Fancier cleaning techniques are the subject of NLP (e.g. removing punctuation, stemming, etc).
                '''
                word = word.lower() #Force the word to be lower case
                yield word, 1  # Mapper outputs a <key, value> pair: in this case, word as key and 1 as value
    
    def reducer(self, word, counts): #key=word, value=counts
        total = 0 
        for i in counts: # counts is an iterator. Since the mapper yielded 1 values, it consists of a series of 1's.
            total += i 
        output = word + '\t' + str(total)
        yield '', output
        
if __name__ == "__main__": 
    TwitterWordCount.run()
````


```
$ python twitter_wc_mrjob.py data/tweets_raw.json > twitter_wordcount.txt
```

# 3. Word Count for Each Twitter User
What if we want to get separate word counts for each Twitter user?
We do this by creating a composite key. The composite key consists of a user-word pair such as:
`('nyctaxi', 'driver')`

So, each line in the output would look something like,
`('nyctaxi', 'driver') 1`

## How to do it in pure python:

````python
word_dict_by_screenname = {}
for tweet in all_tweets:
    sname= tweet['user']['screen_name'] # The screen name of the user who created the tweet.
    text = tweet['text'] 
    words = text.split()
    for w in words:
        w=w.lower()
        dict_key = (sname, w) # Instead of simply taking the word as key, we create a tuple consisting of user screen name and word
        if dict_key not in word_dict_by_screenname:
            word_dict_by_screenname[dict_key] = 1
        else:
            word_dict_by_screenname[dict_key] += 1
word_dict_by_screenname 
````

## The MapReduce way
Note that now the output key of the mapper function is a tuple (screen_name, word).

````python
## Again, we import the MRJob class which is the base class of the mrjob library
import json
import mrjob
from mrjob.job import MRJob

class UserWordCount(MRJob):
    INPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #RawValueProtocol takes the entire line as the input value to the mapper. Assumes no key
    OUTPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #In the reducer, ignores the output key and only writes the value to file
    
    def mapper(self, _, line): 
        tweet = json.loads(line)
        screen_name = tweet['user']['screen_name']
        text = tweet['text'] # This is the tweet content (e.g. '@NeilECollins @nyc311 Thank you for reading, Mr. Collins.  Having a TLC driver license is a privilege.')
        
        ## Basic tokenization: Split the tweet text into words by whitespace
        words = text.split() #['@NeilECollins', '@nyc311', 'Thank', 'you', 'for', 'reading,', 'Mr.', 'Collins.', 'Having', 'a', 'TLC', 'driver', 'license', 'is', 'a', 'privilege.']
        for word in words:
            if word[0] != '@' and word[:4] != "http": # Filter out @mentions (e.g. @nyc311) and links (e.g. https://t.co....)
                word = word.lower()
                
                yield (screen_name, word), 1  #Note that the output key is a tuple of screen name and word.
    
    def reducer(self, key0, counts): #key=(screen_name, word), value=counts
        screen_name, word = key0
        total = 0
        for i in counts: # counts is an iterator. Since the mapper yielded 1 values, it consists of a series of 1's.
            total += i 
        output = screen_name + '\t' + word + '\t' + str(total)
        yield '', output
        
if __name__ == "__main__": 
    UserWordCount.run()
````

Now, run the above from command line:
```
$ python twitter_user_wc_mrjob.py data/tweets_raw.json
```

# 4. Twitter User Co-Mentions

Instead of counting words, we can count the frequency of two Twitter users mentioned together in tweets. The output would look like the following:

> user_A | user_B | comention_frequency

- user_A and user_B were mentioned together in different tweets X number of times.




## In Pure Python (without MapReduce)

````python
## First, let's see how the mentioned user information can be accessed in the data:
for a_tweet in all_tweets:
    mentioned = a_tweet['entities']['user_mentions']
    if len(mentioned)>2:
        break

## In this example, three users (party_kong, NYPDnews, and AndyKingNYC) were mentioned together in a tweet
mentioned
````

````python
## Create a dictionary where key is a pair of co-mentioned users and value is the frequency 
import itertools

co_mentions = {}
for tweet in all_tweets:
    mentioned = tweet['entities']['user_mentions']
    
    # Only consider tweets where more than one user was mentioned.
    if len(mentioned) > 1:
        # let's just extract the screen names and put them in a list
        m_snames=[]
        for each_user in mentioned: 
            m_snames.append( each_user['screen_name'] ) #parse the screen name from the data structure and append it to a list of screen names (m_snames)
        
        # We use the itertools library to create pairs of comentioned users
        for user_pair in itertools.combinations(m_snames, 2): # Generate pairs of users and iterate through those pairs
            if user_pair not in co_mentions:
                co_mentions[ user_pair ] = 1
            else:
                co_mentions[ user_pair ] += 1

# Inspect output
print("Number of co-mentioned pairs:", len(co_mentions))
for i, j in co_mentions.items():
    print("example pair", i, j)
    break # Break out of the loop since one item is sufficient for inspection
````

## The MapReduce Way

````python
import mrjob
from mrjob.job import MRJob
import json
import itertools

class CoMentions(MRJob):
    INPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #RawValueProtocol takes the entire line as the input value to the mapper. Assumes no key
    OUTPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #In the reducer, ignores the output key and only writes the value to file
    
    def mapper(self, _, line): 
        tweet = json.loads(line)
        mentioned = tweet['entities']['user_mentions']
        m_snames=[] #List of the screen names of those who were mentioned
        
        if len(mentioned) > 1: # If more than one user was mentioned in this tweet
            for each_user in mentioned: 
                m_snames.append( each_user['screen_name'] ) #parse the screen names from the data structure and put them in a list of screen names (m_snames)
            
            for user_pair in itertools.combinations(m_snames, 2): # Generate pairs of users and iterate through those pairs
                ## itertools.combinations('ABCD', 2) --> AB AC AD BC BD CD
                ordered_pair = ( max(user_pair), min(user_pair) )
                yield ordered_pair, 1 #Sort the pair so that same pairs end up in the same reducer
                
    def reducer(self, user_pair, counts): #key=(screen_name, word), value=counts
        user_A, user_B = user_pair
        total = 0
        for i in counts: # counts is an iterator. Since the mapper yielded 1 values, it consists of a series of 1's.
            total += i 
        output = user_A + '\t' + user_B + '\t' + str(total)
        yield '', output
        
if __name__ == "__main__": 
    CoMentions.run()
````

# 5. Multistep Job: Each User's Most Frequently Co-Mentioned User 
We can create a job that chains multiple map-reduce steps to get an output.
To do so, we write multiple mapper and reducer functions and execute them in one sequence.

That sequence is specified in a function named `step` that makes mrjob chain these mapper and reducer functions in some order.

In the example below, we write a new class `MostCoMentioned` that modifies the `CoMention` class above. The objective is to output three tab-separated columns: (a) a focal user, (b) the focal user's most frequently co-mentioned user, and (c) the number of times he/she was co-mentioned with the focal user. 

````python
import json
import itertools
import mrjob
from mrjob.job import MRJob
from mrjob.step import MRStep # A class that allows you to explicitly specify the mapper and the reducer function to use in a step 


class MostCoMentioned(MRJob):
    INPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #RawValueProtocol takes the entire line as the input value to the mapper. Assumes no key
    OUTPUT_PROTOCOL = mrjob.protocol.RawValueProtocol #In the reducer, ignores the output key and only writes the value to file

    def mapper1(self, _, line): 
        tweet = json.loads(line)
        mentioned = tweet['entities']['user_mentions']
        m_snames=[] #List of the screen names of those who were mentioned
        if len(mentioned) > 1: # If more than one user was mentioned in this tweet
            for each_user in mentioned: 
                m_snames.append( each_user['screen_name'] ) #parse the screen name from the data structure and append it to a list of screen names (m_snames)
            for user_pair in itertools.combinations(m_snames, 2): # Generate pairs of users and iterate through those pairs
                ## itertools.combinations('ABCD', 2) --> AB AC AD BC BD CD
                ordered_pair = ( max(user_pair), min(user_pair) )
                yield ordered_pair, 1 #Sort the pair so that same pairs end up in the same reducer
            
    def reducer1(self, user_pair, counts): #key=(screen_name, word), value=counts
        user_A, user_B = user_pair
        total = sum(counts)
        yield user_A, (user_B, total) #yield one record for user_A: In the next reduce step, all comentioned users of user_A will be grouped
        yield user_B, (user_A, total) #yield one record for user_B: In the next reduce step, all comentioned users of user_B will be grouped
    
    def reducer2(self, focal_user, values1): #values1 = [(user_1, total_1), (user_2, total_2), (user_3, total_3)...]
        max_value=0 #initialize maximum comention value at 0
        for each in values1: # Iterate through each (user, frequency) pair 
            comentioned_user, frequency = each
            if frequency > max_value:
                max_value = frequency #update max_value to the new maximum number of comentions
                max_mentioned = comentioned_user # the new comentioned user with the maximum number of comentions up to this point
        
        ## At the end of the for-loop, we end up with the most frequently co-mentioned user of the focal_user
        
        result_string = focal_user + '\t' + max_mentioned + '\t' + str(max_value)
        yield '', result_string

        
    def steps(self): # Specify the individual mapper and reducer functions to be used 
        first_step = MRStep(mapper=self.mapper1, reducer=self.reducer1) # The first step should execute the above-defined self.mapper and self.reducer1
        second_step = MRStep( reducer=self.reducer2 ) # The output from self.reducer1 is fed straight into self.reducer2 without having to go through a second mapper.
        
        all_steps = [ first_step, second_step ] # Create a list that contains the steps in order
        return all_steps
        
        
if __name__ == "__main__": 
    MostCoMentioned.run() 
````

Now, run the above from command line:
```
$ python twitter_most_comentioned_mrjob.py data/tweets_raw.json
```
